```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### let's load the clean data

```python
df = pd.read_csv('C:/Users/odhia/OneDrive/Desktop/datascienceproject/data/clean_df.csv')
df.head()
```

### distribution plots for the columns

```python
cols = ['mean_radius', 'mean_texture', 'mean_perimeter', 'mean_area',
       'mean_smoothness']
for i in cols:
    sns.histplot(x=df[i])
    plt.title(f'Histplot  (for {i})')
    plt.ylabel(f'{i}')
    plt.show()
    # insight; all the distributions are gaussian
```

```python
for i in cols:
    sns.boxplot(x=df['diagnosis'], y=df[i], data=df)
    plt.title(f'BoxPlot of Diagnosis (for {i})')
    plt.ylabel(f'{i}')
    plt.show()
```

### plot for mean_smoothness against mean_radius,with respect to diagnosis

```python
sns.scatterplot(x = df.mean_smoothness,y = df.mean_radius,hue = df.diagnosis)
# the lower the mean_smoothness and mean radius the more likely person to have a cancer
```

### plot for mean perimeter aginst mean texture,with respect to diagnosis

```python
sns.scatterplot(x = df.mean_texture,y = df.mean_perimeter,hue = df.diagnosis)
# the lower the mean texture and mean perimeter,the more likely a person is to have a cancer
```
